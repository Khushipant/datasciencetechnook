# <b>Problem 4 : You need to clean the data a little bit now. Replace all column values with the values <code>?</code> and <code>n.a</code> with <code>NaN</code>

```python
# Your code here
data.replace(["?","n.a"],np.nan, inplace=True)

```

```python
print(data.isnull().sum())
```

```python
numeric_columns = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
data[numeric_columns] = data[numeric_columns].astype(float)
```

```python

data['price'] = data['price'].astype(float)
```

# <b>Problem 5 : Print the company name which manufactured the most expensive car. Also print the price of the most expensive car

```python
# Your code here
highest = np.argmax(data["price"])
high_price=data.loc[highest,"price"]
high_make=data.loc[highest,"make"]
print(high_price,high_make)
```

# <b>Problem 6 : Calculate maximum horsepowers for each company. </b>

```python
# Your code here
max_power=data.groupby("make")["horsepower"].max()
print(max_power)
```

# <b> Problem 7 : Print the total number of cars manufactured by each company

```python
# Your code here
total_car_bycomp=data["make"].value_counts()
print(total_car_bycomp)
```

# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# Your code here. Use apply and lambda
def updated_price(row):
    if row['engine-location'] == 'front':
        return row['price']
    elif row['engine-location'] == 'rear':
        return row['price'] * 2

data['updated_price']=data.apply(updated_price,axis=1)
print(data)
```

# <b>Problem 9 : Sort the dataframe by car and price columns

```python
# Your code here
cars_data = data.sort_values(by=['make', 'price'])
print(cars_data)
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers.

```python
# Your code here. Use apply
def con_num_of_doors(row):
    if row['num-of-doors'] == 'four':
        return 4
    elif row['num-of-doors'] == 'two':
        return 2
    else:
        return None

data['num-doors-int'] = data.apply(con_num_of_doors, axis=1)

print(data)
```

# <b>Problem 11 : Concatenate the two data frames given below firstly row wise and secondly column wise.</b>

<code>GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}</code>

<code>japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}</code>

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}

japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
# Your code here


df_german = pd.DataFrame(GermanCars)
df_japanese = pd.DataFrame(japaneseCars)

concatenated_row_wise = pd.concat([df_german, df_japanese], ignore_index=True)
concatenated_column_wise = pd.concat([df_german, df_japanese], axis=1)

print("Concatenated Row-Wise:\n",concatenated_row_wise)
print("\nConcatenated Column-Wise:\n",concatenated_column_wise)

```
